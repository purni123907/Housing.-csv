```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Dell\OneDrive\Desktop\ML exercise\Housing.csv")
```

```python
print(df.head())
```

```python
Q1 = df.quantile(0.25, numeric_only=True)
Q3 = df.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1
```

```python
num_df = df.select_dtypes(include='number')

Q1 = num_df.quantile(0.25)
Q3 = num_df.quantile(0.75)
IQR = Q3 - Q1

num_df = num_df[~((num_df < (Q1 - 1.5 * IQR)) |
                  (num_df > (Q3 + 1.5 * IQR))).any(axis=1)]

print("shape:", num_df.shape)
```

```python
num_df = df.select_dtypes(include='number')

Q1 = num_df.quantile(0.25)
Q3 = num_df.quantile(0.75)
IQR = Q3 - Q1

mask = ~((num_df < (Q1 - 1.5 * IQR)) |
         (num_df > (Q3 + 1.5 * IQR))).any(axis=1)

df_clean = df[mask]

print("original:", df.shape)
print("clean:", df_clean.shape)
```
